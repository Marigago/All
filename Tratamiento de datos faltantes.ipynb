# Imputación de datos

```python
#@title Sustitución por media
m=df.mean()
m
```

```python
df_m=df.fillna(m)
df_m.loc[nansA,'A']
```

```python
# La media aritmética no se modifica
m-df_m.mean()
```

```python
# La disperción se contrae
df.std()-df_m.std() # Vemos solo valores positivos
```

```python
#@title Sustitución por mediana
me=df.median()
me
```

```python
df_me=df.fillna(me)
df_me.loc[nansA,'A']
```

```python
# La mediana no se modifica
me-df_me.median()
```

```python
# La varianza se reduce
df.var()-df_me.var()
```

```python
#@title Sustitución por moda
mo=df.mode()
mo.iloc[0] # Escogemos el primero
```

```python
df_mo=df.fillna(mo.loc[0])
df_mo.loc[nansA,'A']
```

```python
# La moda no se modifica
mo.loc[0]-df_mo.mode().loc[0]
```

```python
df_mo.mode()
```

```python
# En el caso de variables numéricas continuas, el sesgo se hace más negativo
df.skew()-df_mo.skew()
```

```python
H=df.hist()
```

```python
H=df_mo.hist()
```

```python
#@title Sustitución aleatoria
#
# Escogemos un valor aleatorio de 'A' y lo sustituimos en todos los NaNs
vaA=np.random.choice(df.A.dropna())
vaA
```

```python
df_random=df.copy()
df_random['A']=df.A.fillna(vaA)
df_random.loc[nansA,'A']
```

```python
df_random=df.copy()
for var in df:
  val=np.random.choice(df[var].dropna())
  df_random[var]=df[var].fillna(val)
df_random.isna().sum() # No hay valores perdidos :D
```

```python
# Sustitución de un valor aleatorio por cada NaN
#
# Identificamos onde están los valores perdidos de 'A'
np.where(df.A.isna())[0]
```
