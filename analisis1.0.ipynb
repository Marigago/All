**NORMALIZACIÓN DE TODOS LOS DATOS NUMERICOS**

```python
from sklearn.preprocessing import MinMaxScaler
from scipy import stats
for i in variables_numericas:
    scaler = MinMaxScaler()
    df[i] = scaler.fit_transform(df[[i]])
    print(df[i].skew())
    print(df[i].hist(bins=50))
```

Dado que este trabajo no tiene un objetivo especidfico sobre analisis de correlaciones, que es quizas donde aplica analizar los identificadores y si eso influye en otra variable, los descartamos y solo analizamos las variable que sí tienen variación en el contenido de sus filas.

```python
for i in variables:
    print("variable: \n", i, "\n", "\n", df[i].describe(), "\n")
```

```python
variable_lebel_manual=["Age Group", "Post Type",  "Audience Gender", "Weekday Type"]
for i in variable_lebel_manual: print(df[i].value_counts(), "\n")
```

Manualmente aplicamos lebel encoding para las variables que tienen menos de 3 valores en la columna.

```python
df["Age Group"] = df["Age Group"].apply(lambda x:{"Senior Adults": 1, "Mature Adults": 2, "Adolescent Adults": 3}.get(x))
df["Post Type"] = df["Post Type"].apply(lambda x:{"Video": 1, "Link": 2, "Image": 3}.get(x))
df["Audience Gender"] = df["Audience Gender"].apply(lambda x:{"Male": 1, "Female": 2, "Other": 3}.get(x))
df["Weekday Type"] = df["Weekday Type"].apply(lambda x:{"Weekday": 0, "Weekend": 1}.get(x))
```

```python
for i in variable_lebel_manual:
    print("variable: \n", i, "\n", "\n", df[i].describe(), "\n")
```

```python
variables_encod = list(set(variables) - set(variable_lebel_manual))
variables_encod.remove("Post ID")
variables_encod
```

```python
for i in variables_encod:
    print("variable: \n", i, "\n", "\n", df[i].describe(), "\n")
```

Para columnas con mas de 3 valores en ella, le aplicamos lebel encoding de sklearn

```python
from sklearn.preprocessing import LabelEncoder
for i in variables_encod:

    label_encoder = LabelEncoder()

    # Aplicar Label Encoding a la columna 'Columna'
    df[i] = label_encoder.fit_transform(df[i])

for i in variables_encod:
    print("variable: \n", i, "\n", "\n", df[i].describe(), "\n")
```
